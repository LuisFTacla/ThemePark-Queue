```python
import unicodedata
import pandas as pd
from pathlib import Path
import glob
import os
```

```python
REPO_ANTIGO = Path("../bcw-queue")
MAPA_ATRACOES = {
    "REBULICO": (11444, "Rebuliço"),
    "SPIN BLAST": (12325, "Spin Blast"),
    "SUPER SOAKER SPLASH": (12326, "Super Soaker Splash"),
    "XICARAS MALUCAS": (11373, "Xícaras Malucas")
}
```

## FUNÇÕES AUXILIARES

```python
def normalizar_texto(texto):
    if pd.isna(texto): return ""
    # Remove acentos, cedilhas e coloca em maiúsculas
    nksf = unicodedata.normalize('NFKD', str(texto))
    texto_sem_acentos = "".join([c for c in nksf if not unicodedata.combining(c)])
    return texto_sem_acentos.upper().strip()
```

```python
def remover_acentos(txt):
    if not isinstance(txt, str): return txt
    return "".join(c for c in unicodedata.normalize('NFD', txt)
                   if unicodedata.category(c) != 'Mn')
```

## FONTE 1

```python
def preprocess_recovered_data_1():
		path = REPO_ANTIGO / "recovered_data/bcw_src1"
		all_files = glob.glob(os.path.join(path, "*.csv"))
		df_list = []
		for filename in all_files:
				df = pd.read_csv(filename)
				df['source_file'] = os.path.basename(filename)
				df_list.append(df)
		full_df = pd.concat(df_list, ignore_index=True)
		full_df['ride_norm'] = full_df['Ride'].apply(normalizar_texto)
		full_df['id'] = full_df['ride_norm'].map(lambda x: MAPA_ATRACOES.get(x, (None, None))[0])
		full_df['name'] = full_df['ride_norm'].map(lambda x: MAPA_ATRACOES.get(x, (None, None))[1])
		final_df = full_df[full_df['id'].isin([11444, 12325, 12326, 11373])].copy()
		final_df['Date/Time'] = pd.to_datetime(final_df['Date/Time'])
		final_df['data_local'] = final_df['Date/Time'].dt.date
		final_df['hora_local'] = final_df['Date/Time'].dt.time
		final_df['is_open'] = True 
		final_df['wait_time'] = final_df['Wait Time']
		return final_df[['id', 'name', 'is_open', 'wait_time', 'data_local', 'hora_local']]
```

```python
def normalizar_marco(df_marco, mapa_ids, dicionario_sinonimos):
    novo_df = df_marco.copy()
    
    # --- AJUSTE 1: Limpeza e Padronização de nomes ---
    # Primeiro, aplicamos o dicionário de sinônimos (ex: 4D vira 2D)
    novo_df['ride_name'] = novo_df['ride_name'].replace(dicionario_sinonimos)
    
    # --- AJUSTE 2: Fallback para nomes sem acento ---
    # Se o nome ainda não estiver no mapa_ids, tentamos casar removendo acentos
    # Criamos um mapa auxiliar sem acentos para comparação
    mapa_sem_acentos = {remover_acentos(k): v for k, v in mapa_ids.items()}
    
    def buscar_id(nome):
        # 1. Tenta busca direta
        if nome in mapa_ids:
            return mapa_ids[nome]
        # 2. Tenta busca sem acento
        nome_limpo = remover_acentos(nome)
        if nome_limpo in mapa_sem_acentos:
            return mapa_sem_acentos[nome_limpo]
        return None

    novo_df['id'] = novo_df['ride_name'].apply(buscar_id)
    
    # 3. Criar a coluna is_open
    novo_df['is_open'] = novo_df['wait_time'].apply(lambda x: True if x >= 0 else False)
    
    # 4. Renomear e converter tipos (ajustado para tratar formatos de hora variados)
    novo_df = novo_df.rename(columns={
        'ride_name': 'name',
        'date': 'data_local',
        'time': 'hora_local'
    })
    
    novo_df['data_local'] = pd.to_datetime(novo_df['data_local']).dt.date
    # Usamos errors='coerce' para não quebrar se houver lixo na coluna de hora
    novo_df['hora_local'] = pd.to_datetime(novo_df['hora_local'], format='%H:%M:%S', errors='coerce').dt.time
    
    colunas_finais = ['id', 'name', 'is_open', 'wait_time', 'data_local', 'hora_local']
    novo_df = novo_df[colunas_finais]
    
    # Relatório de conferência
    linhas_sem_id = novo_df['id'].isna().sum()
    if linhas_sem_id > 0:
        nomes_faltantes = novo_df[novo_df['id'].isna()]['name'].unique()
        print(f"Aviso: {linhas_sem_id} registros sem ID removidos. Nomes: {nomes_faltantes}")
        novo_df = novo_df.dropna(subset=['id'])
    
    novo_df['id'] = novo_df['id'].astype(int)
    return novo_df
```

```python
# Execução
df_marco_normalizado = normalizar_marco(data_25_mar, mapa_ids, sinonimos_rides)
```
